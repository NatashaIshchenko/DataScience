# Numpy - библиотека для работы с числовыми масивами

```python
import numpy as np
```

array(object) — n-мерный массив из любой (возможно, вложенной) последовательности

eye(N, M=N, k=0) - двумерный массив с N строками с единицами на диагонали и нулями во всех остальных позициях. Число столбцов M по умолчанию равно N, k — сдвиг диагонали (0 для основной диагонали, положительные числа для верхних диагоналей и отрицательные для нижних)

zeros(shape) -- новый массив указанной формы, заполненный нулями

ones(shape) -- новый массив указанной формы, заполненный единицами

full(shape, fill_value) -- новый массив указанной формы, заполненный fill_value

выбор элементов из массива np_a[start:stop:step] 

размеры массива np_a.shape
reshape(x,y), где x - строки, y - столбцы

a.flatten() — превращает массив в одномерный

a.T или a.transpose(*axes) — транспонирование (или смена порядка осей в случае, когда размерность массива больше двух)

np.fromiter создаёт NumPy-массив из итератора, то есть заставляет итератор вычислить все доступные значения и сохраняет их в массив; пример: X = np.fromiter(map(int, input().split()), np.int).reshape(x_shape)

=== Математические преобразования ===

np_a.mean(), np_a.max(), np_a.min()

var() - дисперсия

std() - среднеквадратическое отклонение

a.argmin(axis=None), a.argmax(axis=None) — индексы минимального и максимального элемента

a.sum(axis=None), a.prod(axis=None) — сумма и произведение всех элементов вдоль указанной оси

a.cumsum(axis=None), a.cumprod(axis=None) — частичные суммы и произведения (для (a1,⋯,an) вектор частичных сумм — это (a1,a1+a2,⋯,a1+⋯+an))

Линейная алгебра - пакет numpy.linalg 

    a.dot(b) — матричное произведение двух массивов (размерности должны быть согласованы)
    
    linalg.matrix_power(M, n) — возведение матрицы M в степень n
    
    a.T — транспонирование
    
    linalg.norm(a, ord=None) — норма матрицы a, по умолчанию норма Фробениуса для матриц и L2-норма для векторов; 
    
    linalg.inv(a) — матрица, обратная к a (если a необратима, выбрасывается LinAlgError;
    псевдообратная считается через linalg.pinv(a))

```python
# zeros(shape) -- новый массив указанной формы, заполненный нулями

np1 = np.zeros((3,2))
print(np1)
```

```python
# eye(N, M=N, k=0) - двумерный массив с N строками с единицами на диагонали и нулями во всех остальных позициях. Число столбцов M по умолчанию равно N, k — сдвиг диагонали (0 для основной диагонали, положительные числа для верхних диагоналей и отрицательные для нижних)

np1 = np.eye(5,5,1)
np1
```

```python
# ones(shape) -- новый массив указанной формы, заполненный единицами

np1 = np.ones((2,3))
np1
```

```python
# full(shape, fill_value) -- новый массив указанной формы, заполненный fill_value

np1 = np.full((3,2), 4)
np1
```

```python
# arange(x,y,z), где x - начало, y - конец, z - шаг

np2 = np.arange(1,10,2)
print(np2)
```

```python
# linspace(x,y,z), где x - начало, y - конец, z - количество элементов/точек

np3 = np.linspace(1,10,3)
print(np3)
```

```python
np4 = np.arange(12)
print(np4)
```

```python
# reshape(x,y), где x - строки, y - столбцы

np5 = np4.reshape(4,3)
print(np5)
print(np4)
```

```python
# математические операции

a = np.array([1,2,3,4])
b = np.array([5,6,7,8])
print('a+b = ', a+b)
print('a-b = ', a-b)
print('sin(a) = ', np.sin(a))
```

```python
list1 = [1,2,3,4,5,6,4,3,2,33,4,1,2,3,4,5,6,7,8,6,5,4,3,2,1,3,4,5,6,7,8,9,0,99,66,44,33,22,11,21,32,43,54,65,67]
```

```python
# среднее значение
mean_list1 = np.array(list1).mean()
print(mean_list1)
```

```python
# дисперсия - мера разброса значений случайной величины относительно её математического ожидания
variance_list1 = np.array(list1).var()
print(variance_list1)
```

```python
# среднеквадратическое отклонение (корень из дисперсии)
standard_deviation_list1 = np.array(list1).std()
print(standard_deviation_list1)
```

# Считывание данных с файла

"sbux.csv" — имя файла, из которого считываются данные.

usecols — список колонок, которые нужно использовать. Если параметр не указан, считываются все колонки.

skiprows — количество рядов в начале файла, которые нужно пропустить. По умолчанию (если значение параметра не указано явно) skiprows = 0.

delimiter — разделитель столбцов в одной строке, в csv-файлах это запятая, по умолчанию разделителем является любой пробел (в том числе — знак табуляции).

dtype — словарь из названий колонок (переменных) и типов хранящихся в них значений. NumPy использует свою собственную систему типов, и названия именно этих типов нужно указать. По умолчанию функция попытается самостоятельно угадать, какому типу принадлежат подаваемые на вход значения.

```python
sbux = np.loadtxt("sbux.csv", usecols=(0,1,2), skiprows=0, delimiter=",")
print(sbux)
```

```python
from urllib.request import urlopen
f = urlopen('https://stepic.org/media/attachments/lesson/16462/boston_houses.csv')
sbux = np.loadtxt(f, usecols=(0,1,2), skiprows=1, delimiter=",", 
                 dtype={'names': ('medv', 'crim', 'zn'),
                             'formats': ('f4', 'f4', 'f4')})
print(sbux)
```
